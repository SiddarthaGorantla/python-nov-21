```python
def fib(n):
    #base case
    if n==0 or n==1:
        return n
    
    #recursive logic 
    f1 = fib(n-1)
    f2 = fib(n-2)
    return f1 + f2
    
```

```python
fib(6)
```

```python
fib(7)
```

```python
def party(n):
    if n==1:
        return 1
    if n==2:
        return 2
    return party(n-1) + (n-1)*party(n-2)
```

```python
party(4)
```

```python
party(3)
```

```python
def maze(m,n):
    if m==1 and n==1:
        return 1
    if m < 1 or n < 1:
        return 0
    # rec
    top = maze(m-1,n) 
    right = maze(m,n-1)
    return top + right
```

```python
maze(3,3)
```
